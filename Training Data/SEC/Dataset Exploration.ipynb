# SEC EDGAR Dataset

SEC filings included:
- **10-K**: Annual reports
- **10-Q**: Quarterly reports
- **8-K**: Event-driven disclosures


```python
# download all stock ticker symbols from SEC 
# https://www.sec.gov/files/company_tickers.json

import pandas as pd
import json
import os
from sec_edgar_downloader import Downloader
import sys
from tqdm import tqdm  # Added for progress bar
```

```python
tickers_df = pd.read_csv('Q1_tickers.csv')
```

```python

# Set the directory where filings will be downloaded

# Initialize the downloader with the target directory
dl = Downloader(company_name="Reseach for Financial NLP", email_address="tadoko1614@avulos.com")

tickers = tickers_df['ticker']
with tqdm(total=len(tickers), desc="Processing", ncols=80) as pbar:
    for ticker in tickers:
        pbar.set_description(f"Currently working on {ticker}")
        for filing_type in ["10-K", "10-Q", "8-K"]:
            try:
                dl.get(filing_type, ticker)
            except Exception as e:
                print(f"An error occurred for {ticker} ({filing_type}): {e}")
        pbar.update(1)

```
